Escreva um programa que leia um número e verifique se é ou não um número primo. Para
fazer essa verificação, calcule o resto da divisão do número por 2 e depois por todos os
números ímpares até o número lido. Se o resto de uma dessas divisões for igual a zero, o
número não é primo. Observe que 0 e 1 não são primos e que 2 é o único número primo
que é par.

````python
def verificar_primo(numero):
    if numero == 0 or numero == 1:
        return False

    if numero == 2:
        return True

    if numero % 2 == 0:
        return False

    for i in range(3, int(numero**0.5) + 1, 2):
        if numero % i == 0:
            return False

    return True

numero = int(input("Digite um número inteiro: "))

if verificar_primo(numero):
    print(numero, "é um número primo.")
else:
    print(numero, "não é um número primo.")
````

8.Escreva um programa que leia uma string e imprima quantas vezes cada caractere aparece
nessa string.

String: TTAAC

Resultado:

T: 2x

A: 2x

C: 1x

````python
def contar_caracteres(string):
    contagem = {}

    for char in string:
        if char in contagem:
            contagem[char] += 1
        else:
            contagem[char] = 1

    return contagem

def imprimir_contagem(contagem):
    for char, count in contagem.items():
        print(f"{char}: {count}x")

string = input("Digite uma string: ")
resultado = contar_caracteres(string)
imprimir_contagem(resultado)
````

9.Escreva um jogo da velha para dois jogadores. O jogo deve perguntar onde você quer jogar
e alterar entre os jogadores. A cada jogada, verifique se a posição está livre. Verifique
também quando um jogador venceu a partida. Um jogo da velha pode ser visto como uma
lista de 3 elementos, na qual cada elemento é outra lista, também com três elementos.

Exemplo do jogo:

image.png

Em que cada posição pode ser vista como um número. Confira a seguir um exemplo das
posições mapeadas para a mesma posição de seu teclado número.

image.png

````python
def criar_tabuleiro():
    tabuleiro = [[' ', ' ', ' '], [' ', ' ', ' '], [' ', ' ', ' ']]
    return tabuleiro

def imprimir_tabuleiro(tabuleiro):
    print('---------')
    for linha in tabuleiro:
        print(f'| {linha[0]} | {linha[1]} | {linha[2]} |')
        print('---------')

def jogada_valida(linha, coluna, tabuleiro):
    if linha < 0 or linha > 2 or coluna < 0 or coluna > 2:
        return False
    if tabuleiro[linha][coluna] != ' ':
        return False
    return True

def fazer_jogada(linha, coluna, jogador, tabuleiro):
    tabuleiro[linha][coluna] = jogador

def verificar_vitoria(jogador, tabuleiro):
    # Verificar linhas
    for linha in tabuleiro:
        if linha.count(jogador) == 3:
            return True

    # Verificar colunas
    for coluna in range(3):
        if tabuleiro[0][coluna] == jogador and tabuleiro[1][coluna] == jogador and tabuleiro[2][coluna] == jogador:
            return True

    # Verificar diagonais
    if tabuleiro[0][0] == jogador and tabuleiro[1][1] == jogador and tabuleiro[2][2] == jogador:
        return True
    if tabuleiro[0][2] == jogador and tabuleiro[1][1] == jogador and tabuleiro[2][0] == jogador:
        return True

    return False

def jogar_jogo_da_velha():
    tabuleiro = criar_tabuleiro()
    jogador_atual = 'X'

    while True:
        imprimir_tabuleiro(tabuleiro)
        print(f"Jogador {jogador_atual}, é sua vez.")
        posicao = input("Digite a posição para jogar (1-9): ")

        if len(posicao) != 1 or not posicao.isdigit():
            print("Posição inválida. Digite um número de 1 a 9.")
            continue

        posicao = int(posicao)
        linha = (posicao - 1) // 3
        coluna = (posicao - 1) % 3

        if not jogada_valida(linha, coluna, tabuleiro):
            print("Posição inválida. Escolha uma posição vazia.")
            continue

        fazer_jogada(linha, coluna, jogador_atual, tabuleiro)

        if verificar_vitoria(jogador_atual, tabuleiro):
            imprimir_tabuleiro(tabuleiro)
            print(f"Jogador {jogador_atual} venceu!")
            break

        if ' ' not in [pos for linha in tabuleiro for pos in linha]:
            imprimir_tabuleiro(tabuleiro)
            print("O jogo empatou!")
            break

        jogador_atual = 'O' if jogador_atual == 'X' else 'X'


# Executar o jogo
jogar_jogo_da_velha()
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
X_poly = np.linspace(-3, 5, 81)
print(X_poly[:5], '...', X_poly[-5:])

def cost_function(X):
  return X * (X-2)

y_poly = cost_function(X_poly)
plt.plot(X_poly, y_poly)
plt.xlabel('Parameter value')
plt.ylabel('Cost function')
plt.title('Error surface')
````

````python
def gradient(X):
  return (2*X) - 2
x_start = 4.5
learning_rate = 0.75
x_next = x_start - gradient(x_start)*learning_rate
x_next
````

````python
plt.plot(X_poly, y_poly)
plt.plot([x_start, x_next], [cost_function(x_start), cost_function(x_next)], '-o')
plt.xlabel('Parameter value')
plt.ylabel('Cost function')
plt.legend(['Error surface', 'Gradient descent path'])
````

````python
iterations = 15
x_path = np.empty(iterations,)
x_path[0] = x_start
for iteration_count in range(1, iterations):
  derivative = gradient(x_path[iteration_count-1])
  x_path[iteration_count] = x_path[iteration_count-1] - (derivative*learning_rate)
x_path
````

````python
plt.plot(X_poly, y_poly)
plt.plot(x_path, cost_function(x_path), '-o')
plt.xlabel('Parameter value')
plt.ylabel('Cost function')
plt.legend(['Error surface', 'Gradient descent path'])
````

````python
from sklearn.metrics import roc_curve
````

````python
def cross_val_C_search(k_folds, C_vals, model, X, Y):
  n_folds = k_folds.n_splits
  cv_train_roc_auc = np.empty((n_folds, len(C_vals)))
  cv_test_roc_auc = np.empty((n_folds, len(C_vals)))
  cv_test_roc = [[]] * len(C_vals)
  for c_val_counter in range(len(C_vals)):
    # Define o valor de C para o objeto de modelo
    model.C = C_vals[c_val_counter]
    #Conta os folds para cada valor de C
    fold_counter = 0
    for train_index, test_index in k_folds.split(X, Y):
      X_cv_train, X_cv_test = X[train_index], X[test_index]
      y_cv_train, y_cv_test = Y[train_index], Y[test_index]
      model.fit(X_cv_train, y_cv_train)
      y_cv_train_predict_proba = model.predict_proba(X_cv_train)
      cv_train_roc_auc[fold_counter, c_val_counter] = \
      roc_auc_score(y_cv_train, y_cv_train_predict_proba[:,1])
      y_cv_test_predict_proba = model.predict_proba(X_cv_test)
      cv_test_roc_auc[fold_counter, c_val_counter] = \
      roc_auc_score(y_cv_test, y_cv_test_predict_proba[:,1])
      this_fold_roc = roc_curve(y_cv_test, y_cv_test_predict_proba[:,1])
      cv_test_roc[c_val_counter].append(this_fold_roc)
    fold_counter += 1
    print('Done with C = {}'.format(lr_syn.C))
  return cv_train_roc_auc, cv_test_roc_auc, cv_test_roc
````

````python
from sklearn.model_selection import KFold

n_folds = 4
k_folds = KFold(n_splits=n_folds, shuffle=False)

cv_train_roc_auc, cv_test_roc_auc, cv_test_roc = \
cross_val_C_search(k_folds, C_vals, lr_syn, X_syn_train, y_syn_train)
````
